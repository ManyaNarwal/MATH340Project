```r
nutrients <- read.csv("nutrients_csvfile.csv")
nutrients <- nutrients %>% select(-Category, -Fat, -Sat.Fat)
head(nutrients)
```

```r
# removed coffee, butter, mushrooms
# added oranges
list <- c("Cows' milk", "Apples, raw", "Banana", "Oranges 3in diameter", "Tomatoes", 
                     "Onions", "Bread, cracked wheat", "Eggs raw",
                     "Cheese", "Beef", "chicken", "Salmon", 
                     "Carrots", "Bacon", "Whole-wheat", "Potatoes, baked")
selected_items <- nutrients %>% 
                filter(Food %in% list) %>% 
                select(-Measure) %>%
                slice(-c(15,17))
selected_items
```

```r
selected_items$Protein[selected_items$Protein == "t"] <- 0
selected_items$Fiber[selected_items$Fiber == "t"] <- 0
selected_items$Carbs[selected_items$Carbs == "t"] <- 0
selected_items$Food[selected_items$Food == "chicken"] <- "Chicken"
selected_items$Food[selected_items$Food == "Beef"] <- "Ground beef"
selected_items$Food[selected_items$Food == "Cows' milk"] <- "Partly skimmed milk"
selected_items$Food[selected_items$Food == "Apples, raw"] <- "Apples"
selected_items$Food[selected_items$Food == "Eggs raw"] <- "Eggs"
selected_items$Food[selected_items$Food == "Whole-wheat"] <- "Bread"
selected_items$Food[selected_items$Food == "Cheese"] <- "Processed cheese slices"
selected_items$Food[selected_items$Food == "Salmon"] <- "Canned salmon"
selected_items$Food[selected_items$Food == "Potatoes, baked"] <- "Potatoes"
selected_items$Food[selected_items$Food == "Banana"] <- "Bananas"
selected_items$Food[selected_items$Food == "Tomatoes"] <- "Canned tomatoes"
selected_items$Food[selected_items$Food == "Oranges 3in diameter"] <- "Oranges"
selected_items$Calories[selected_items$Calories == "1,100"] <- 1100
selected_items %>% mutate(Grams = as.numeric(Grams), Calories = as.numeric(Calories), Fiber = as.numeric(Fiber), Carbs = as.numeric(Carbs))
```

```r
print(selected_items)
```

```r
write.csv(selected_items, "selected_items.csv")
```
